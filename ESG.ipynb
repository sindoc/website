```python
import datacommons as dc

# Replace with the actual dataset ID
dataset_id = "dc/kg/m/CO2_emissions"

# Get top 5 countries by CO2 emissions
top_countries_query = f"""
SELECT Country, SUM(CO2_emissions) AS Total_CO2
FROM {dataset_id}
GROUP BY Country
ORDER BY Total_CO2 DESC
LIMIT 5
"""
top_countries_results = dc.query(top_countries_query)

# Get CO2 emissions by industry for each top country
for country in top_countries_results:
    country_query = f"""
    SELECT Industry, SUM(CO2_emissions) AS Industry_CO2
    FROM {dataset_id}
    WHERE Country = '{country["Country"]}'
    GROUP BY Industry
    ORDER BY Industry_CO2 DESC
    """
    country_results = dc.query(country_query)
    print(f"Top CO2 emitting industries in {country['Country']}:\n")
    for industry in country_results:
        print(f"  {industry['Industry']}: {industry['Industry_CO2']}")
```

```python
# Step 1: Install Required Libraries
# If you haven't installed the necessary packages yet, uncomment and run the following lines:
# !pip install datacommons_pandas pandas matplotlib

# Step 2: Import Libraries
import datacommons_pandas as dcpd
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Define the Region and Variables for Emissions Data
# In this case, we will focus on the USA and the sectors we are interested in: Energy, Transportation, and Industry.

geo_id = 'country/USA'  # GeoID for the USA

# Variables related to CO2 emissions by sector
variables = [
    'Emissions_CarbonDioxide_Energy',        # CO2 emissions from the energy sector
    'Emissions_CarbonDioxide_Transportation',# CO2 emissions from the transportation sector
    'Emissions_CarbonDioxide_Industry'       # CO2 emissions from the industry sector
]

# Step 4: Fetch data from Data Commons using the API
df = dcpd.build_multivariate_dataframe([geo_id], variables)

# Step 5: Rename Columns for Clarity
df.columns = ['Energy_CO2_Emissions', 'Transportation_CO2_Emissions', 'Industry_CO2_Emissions']

# Step 6: Convert Data Types and Handle Missing Values
df = df.apply(pd.to_numeric, errors='coerce').dropna()

# Step 7: Display the Data
print("CO2 Emissions Data by Sector in the USA:")
print(df.head())  # Display the first few rows of the data

# Step 8: Plot Total CO2 Emissions by Sector Over Time
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Energy_CO2_Emissions'], label='Energy Sector')
plt.plot(df.index, df['Transportation_CO2_Emissions'], label='Transportation Sector')
plt.plot(df.index, df['Industry_CO2_Emissions'], label='Industry Sector')

# Step 9: Customize the Plot
plt.title('Total CO2 Emissions by Sector in the USA (Annual)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Million Metric Tons)')
plt.legend()
plt.grid(True)
plt.show()

# Step 10: Analyze the Trends (Optional)
# You can calculate and print the percentage share of emissions by each sector
total_emissions = df.sum(axis=1)
sector_share = df.div(total_emissions, axis=0) * 100

print("Percentage Share of Emissions by Sector:")
print(sector_share.head())  # Show the first few rows of the sector share data
```

```python
# Step 1: Install required libraries (if not already installed)
# !pip install wbdata pandas matplotlib

# Step 2: Import necessary libraries
import wbdata  # Library to access World Bank data
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Step 3: Set the indicator code for SDG-related employment data
# The World Bank uses specific indicator codes. "SI.POV.EMP1" could correspond to employment-related data.
# We are assuming here that "SI.POV.EMP1.AGE--Y_GE25" refers to employment for individuals aged 25+ years.
indicator_code = 'SI.POV.EMP1'

# Step 4: Fetch the data for countries (you can select specific countries, or use 'all' to get global data)
# Fetch data for a specific country (e.g., 'USA') or a list of countries
countries = ['USA', 'GBR', 'IND']  # Example countries (USA, UK, India). You can adjust this list.
data = wbdata.get_dataframe(indicator_code, country=countries, data_date=(datetime(2000, 1, 1), datetime(2021, 1, 1)))

# Step 5: Preview the fetched data
print(data.head())

# Step 6: Plot the data for each country over time
plt.figure(figsize=(10, 6))
for country in countries:
    country_data = data.xs(country, level='country')
    country_data.plot(y=indicator_code, label=country)

# Step 7: Customize the plot
plt.title('Employment Rates for Individuals Aged 25+ Years')
plt.xlabel('Year')
plt.ylabel('Employment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()

# Optional: Save the dataframe to a CSV file if needed
data.to_csv('employment_age_25plus.csv')
```
